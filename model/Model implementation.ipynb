```python
drop_columns = ['url', 'region', 'region_url', 'VIN', 'size','image_url','county','state','lat','long','posting_date','description']
d1 = d1.drop(columns = drop_columns)
```

```python
from sklearn.preprocessing import LabelEncoder
for col in d1.columns:
      le = LabelEncoder()
      le.fit(list(d1[col].astype(str).values))
      d1[col] = le.transform(list(d1[col].astype(str).values))
```

```python
d1.head()
```

```python
year_data = d1['year']


price_data = d1['price']

odometer_data=d1['odometer']

#year = np.array(year_data).reshape(-1,1)

price = np.array(price_data).reshape(-1,1)
#odometer = np.array(odometer_data).reshape(-1,1)
features=np.array([year_data,odometer_data]).T
```

```python
model = LinearRegression()
model.fit(features,price)
```

```python
result=model.predict([[113,104870]])
print(result)
d1.info()
```
